# Data Analysis and Visualization

```python
cancelled_percet = df['is_canceled'].value_counts(normalize=True)
print('Percentage of cancellations')
print(cancelled_percet)

plt.figure(figsize=(5,5))
plt.title('Reservation status count')
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax1=sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Set2' )
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.legend(['Not Cancelled','Cancelled'])
plt.title('Reservation status of different hotels', size = 20)
plt.xlabel('Hotel', size = 14)
plt.ylabel('Number of reservations', size = 14)
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
print('Cancellation percentage of Resort hotels')
print(resort_hotel['is_canceled'].value_counts(normalize=True))
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
print('Cancellation percentage of City hotels')
print(city_hotel['is_canceled'].value_counts(normalize=True))
```

```python
ResortHotel_Price = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
CityHotel_Price = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average daily rate in City and Resort hotel', fontsize=30)
plt.plot(ResortHotel_Price.index, ResortHotel_Price['adr'], label = 'Resort Hotel')
plt.plot(CityHotel_Price.index, CityHotel_Price['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.legend(['Not Cancelled','Cancelled'])
plt.title('Reservation status of different hotels', size = 20)
plt.xlabel('Month', size = 14)
plt.ylabel('Number of reservations', size = 14)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('Average daily rate per month', fontsize=30)
sns.barplot('month', 'adr', data = df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.xlabel('Month', size = 14)
plt.ylabel('Average daily rate', size = 14)
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
```

```python
plt.figure(figsize=(10,10))
plt.title('Top 10 countries with reservation cancellations', size =20)
plt.pie(top_10_countries, autopct = '%.2f', labels = top_10_countries.index)
```

```python
cancelled_data_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_data_adr.reset_index(inplace=True)
cancelled_data_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_data_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_data_adr.reset_index(inplace=True)
not_cancelled_data_adr.sort_values('reservation_status_date', inplace=True)
```

```python
plt.figure(figsize=(20,6))
plt.title('Average daily rate', fontsize=30)
plt.plot(not_cancelled_data_adr['reservation_status_date'], not_cancelled_data_adr['adr'], label = 'Not Cancelled')
plt.plot(cancelled_data_adr['reservation_status_date'], cancelled_data_adr['adr'], label = 'Cancelled')
plt.legend(fontsize = 20)
plt.show()
```

```python
modified_cancelled_data_adr = cancelled_data_adr[(cancelled_data_adr['reservation_status_date']>'2016') & (cancelled_data_adr['reservation_status_date']<'2017-09')]
modified_not_cancelled_data_adr = not_cancelled_data_adr[(not_cancelled_data_adr['reservation_status_date']>'2016') & (not_cancelled_data_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average daily rate', fontsize=30)
plt.plot(modified_not_cancelled_data_adr['reservation_status_date'], modified_not_cancelled_data_adr['adr'], label = 'Not Cancelled')
plt.plot(modified_cancelled_data_adr['reservation_status_date'], modified_cancelled_data_adr['adr'], label = 'Cancelled')
plt.legend(fontsize = 20)
plt.show()
```
